```python
import matplotlib.pyplot as plt

# 设置图像大小
plt.figure(figsize=(4, 3.05))  # 稍微增加高度避免标签重叠
plt.rcParams.update({'font.size': 12})


# 原始数据（额外量子比特）→ 转换为总量子比特（x+2）
x_extra = [0, 1, 2, 3, 4, 5]  # 额外量子比特
y_qr_move = [141, 77, 54, 45, 37, 31]  # 电路深度
x_total = [x + 2 for x in x_extra]  # 转换为总量子比特 (2,3,4,5,6)

# 新增点（总量子比特）
x_qiskit_total = 30  # 28+2
y_qiskit = 24
x_qr_map_total = 2   # 0+2
y_qr_map = 150       # 需高于141

# 绘制QR-MOVE折线
plt.plot(x_extra, y_qr_move, "b-", linewidth=2.5, label="QR-MOVE", zorder=5)

# 添加离散点（使用原始x坐标值，但显示总量子比特标签）
plt.scatter([28], [y_qiskit], s=100, c="pink", edgecolor='k', zorder=15, label="Qiskit")
plt.scatter([0], [y_qr_map], s=100, c="green", edgecolor='k', zorder=15, label="QR-Map")

# 虚线连接Qiskit和QR-Map
plt.plot([0, 28], [y_qr_map, y_qiskit], "k--", linewidth=1.2, alpha=0.3)

# 优化箭头（调整位置适应新坐标系）
plt.annotate(
    "",
    xy=(3, 65),       # 终点 (总量子比特≈5)
    xytext=(9, 100), # 起点 (总量子比特≈25)
    arrowprops=dict(arrowstyle="->", lw=2.5, color="darkred", alpha=0.9),
    zorder=8
)
# plt.text(6, 88, "Optimization", 
#          rotation=-39, 
#          color="darkred", 
#          fontsize=12,
#          bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

# 自定义x轴：位置用原始值，标签显示总量子比特
plt.xticks(
    ticks=[0, 1, 2, 3, 4, 5, 28],
    labels=['2', '3', '4', '5', '6', '7', '30'],  # 显示总量子比特
    fontsize=10
)

# 标签和标题
plt.xlabel("Total qubit num", fontsize=13, labelpad=8)
plt.ylabel("Execution time", fontsize=12.5, labelpad=8)
# plt.title("Circuit Depth vs Qubit Count", fontsize=11, pad=12)

# 图例（避免覆盖数据）
plt.legend(loc='upper right', fontsize=11, framealpha=0.95)

# 网格和边界优化
plt.grid(linestyle='--', alpha=0.3)
plt.xlim(-2, 32)  # 增加右侧空间
plt.ylim(15, 160)

# 紧凑布局
plt.tight_layout(pad=1.5)
plt.savefig("width_depth.pdf", bbox_inches='tight')
plt.show()
```
