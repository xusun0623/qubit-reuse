```python
# 导入必要的库
from qiskit.providers.fake_provider import FakeProvider
from qiskit.providers.fake_provider import FakeBackend
import warnings
warnings.filterwarnings('ignore')

# 列出所有可用的模拟后端
fake_provider = FakeProvider()
backends = fake_provider.backends()

print("可用的模拟后端:")
for backend in backends[:10]:  # 显示前10个后端
    print(f"- {backend.name}")

# 选择一个可用的后端（例如第一个）
if backends:
    backend = backends[0]
    print(f"\n使用后端: {backend.name}")
    
    # 获取后端的属性信息
    properties = backend.properties()
    
    # 打印单量子比特测量重置时间
    print("\n=== 单量子比特测量重置时间 ===")
    if properties and hasattr(properties, 'qubits'):
        for qubit_idx in range(min(5, len(properties.qubits))):  # 显示前5个量子比特
            try:
                # 查找readout_length参数（测量重置时间）
                readout_length_params = [param for param in properties.qubits[qubit_idx] 
                                       if param.name == 'readout_length']
                if readout_length_params:
                    readout_reset_time = readout_length_params[0].value
                    unit = readout_length_params[0].unit
                    if unit == 'us':
                        print(f"量子比特 {qubit_idx}: {readout_reset_time:.2f} μs")
                    elif unit == 'ns':
                        print(f"量子比特 {qubit_idx}: {readout_reset_time:.2f} ns")
                    else:
                        print(f"量子比特 {qubit_idx}: {readout_reset_time} {unit}")
                else:
                    print(f"量子比特 {qubit_idx}: 无readout_length数据")
            except Exception as e:
                print(f"量子比特 {qubit_idx}: 无法获取数据")
    
    # 打印双量子比特门(CX门)时间
    print("\n=== 双量子比特门时间 ===")
    if properties and hasattr(properties, 'gates'):
        gates = properties.gates
        cx_gates = [gate for gate in gates if gate.gate == 'cx']
        
        # 显示前10个CX门的时间
        for i, gate in enumerate(cx_gates[:10]):
            if len(gate.parameters) > 0:
                gate_time_param = [param for param in gate.parameters if param.name == 'gate_time']
                if gate_time_param:
                    gate_time = gate_time_param[0].value
                    unit = gate_time_param[0].unit
                    qubits = gate.qubits
                    if unit == 'ns':
                        print(f"CX门 {qubits[0]}-{qubits[1]}: {gate_time:.2f} ns")
                    elif unit == 'us':
                        print(f"CX门 {qubits[0]}-{qubits[1]}: {gate_time:.2f} μs")
                    else:
                        print(f"CX门 {qubits[0]}-{qubits[1]}: {gate_time} {unit}")
```
